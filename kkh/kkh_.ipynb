```python
def to_df(table_name_list='US_UNEMPLOYMENT_RATE', start_date=None, end_date=None, rule='', bool_=False):
    engine = create_engine("oracle+cx_oracle://Ai:0000@192.168.0.48:1521/XE")
    con = engine.connect()
    df = pd.DataFrame()
    
    if type(table_name_list)==str:
        temp_list = []
        temp_list.append(table_name_list)
        table_name_list = temp_list
    
    for table_name in table_name_list:
        query = f"""select * from {table_name}"""
        temp = pd.read_sql_query(text(query), con=con)
        temp = temp.drop_duplicates(temp.columns[0], keep='first')
        temp = temp.set_index(temp.columns[0])
        temp.index.name = ''
        temp.columns = [table_name]

        if rule!='':
            temp = temp.resample(rule=rule).mean()

        df = pd.concat([df, temp], axis=1)

    con.close()
    engine.dispose()
    
    df = df.loc[start_date:end_date].sort_index().dropna()
    if bool_==True:
        df[df.columns] = False
    
    return df

to_df(['US_UNEMPLOYMENT_RATE'], start_date='2008-07-01', end_date='2022-08-10')
```

```python
def to_index(df):
    temp = df.iloc[0]
    
    return df/temp*100
```

```python
def get_end_month(df):
    df = df.sort_index()
    df['year'] = df.index.year
    df['month'] = df.index.month
    df = df.drop_duplicates(subset=['year', 'month'], keep='last')
    df = df.drop(['year', 'month'], axis=1)
    
    return df
```

```python
def get_bool(df):
    df = df.sort_index()
    for col in df.columns:
        df[col] = np.where(df[col]>=0, True, False)
        
    return df
```

```python
def get_date_ago_year(date_):
    if type(date_)==str:
        date_ = str(int(date_[:4])-1)+date_[4:]
    if type(date_)==pd._libs.tslibs.timestamps.Timestamp:
        date_ = pd.Timestamp(year=date_.year-1, month=date_.month, day=date_.day)
    if type(date_)==datetime:
        date_ = date_ - relativedelta(years=1)
    return date_

get_date_ago_year('2010-01-01')
```

```python
def get_benchmark(benchmark_dict={'VTI':0.6, 'BND':0.4}, start_date=None, end_date=None, stats=False):    
    RMTEST = bt.Strategy(
        ''.join([k+str(int(v)) for k, v in benchmark_dict.items()]),
        algos = [
            bt.algos.RunMonthly(
                run_on_first_date=False,
                run_on_end_of_period=True,
                run_on_last_date=True
            ),
            # bt.algos.PrintInfo('{name}:{now}. Value:{_value:0.0f}, Price:{_price:0.4f}'),
            # bt.algos.SelectAll(),
            bt.algos.SelectAll(), 
            bt.algos.WeighSpecified(**benchmark_dict),
            # bt.algos.PrintTempData(),
            bt.algos.Rebalance()
        ]
    )
    
    t = bt.Backtest(RMTEST, to_df(benchmark_dict.keys(), start_date=start_date, end_date=end_date))
    report = bt.run(t)
    
    if stats==True:
        res = report.stats
    else:
        res = report.prices
    
    return res

start_date = '2014-01-01'
end_date = '2022-12-31'
benchmark = get_benchmark(benchmark_dict={'SPY':1}, start_date=start_date, end_date=end_date)
benchmark
```

잼룰
S&P(SPY)가 미국단기채(US_T_BILLS)보다 크면
	1-1. S&P(SPY)가 세계전체주식(VEU)보다 크면
		- S&P(SPY) 사라

	1-2. S&P(SPY)가 세계전체주식(VEU)보다 작으면
		- 세계전체주식(VEU) 사라

S&P(SPY)가 미국단기채(US_T_BILLS)보다 작으면
	2. 미국혼합채(AGG) 사라

0 SHY 사라  
1 S&P 사라  
2 SCZ 사라  
3 AGG 사라  

```python
def original_dual_momentum(strategy_name='OriginalDualMomentum', start_date=None, end_date=None, plot=True, html=False, benchmark=None):
    # 시그널 핵심 알고리즘
    df_m = to_df(['US_T_BILLS', 'SPY', 'ACWI'], start_date=get_date_ago_year(start_date), end_date=end_date)
    df_m = get_end_month(df_m)
    df_m = df_m.pct_change(12).dropna()

    signal_list = []
    for idx, row in df_m.iterrows():
        if (row['SPY']<0)&(row['US_T_BILLS']<0):
            # print(idx, row.values, '현금보유') # 0
            signal_list.append(0)
        else:
            if row['SPY']>row['US_T_BILLS']: # 잼룰
                if row['SPY']>=row['ACWI']:
                    # print(idx, row.values, 'S&P(SPY) 사라') # 1
                    signal_list.append(1)
                else:
                    # print(idx, row.values, '미국을뺀 세계전체주식(ACWI) 사라') # 2
                    signal_list.append(2)
            else:
                # print(idx, row.values, '미국혼합채(AGG) 사라') # 3
                signal_list.append(3)
    df_m['signal'] = signal_list
    
    # 값이 True False인 데이터 프레임
    df_bool = to_df(['SHY', 'SPY', 'ACWI', 'AGG'], start_date=start_date, end_date=end_date, bool_=True)
    df_bool = pd.concat([df_bool, df_m['signal']], axis=1)
    df_bool['signal'] = df_bool['signal'].ffill().bfill().astype(int)
    
    df_bool['SHY'] = np.where(df_bool['signal']==0, True, False)
    df_bool['SPY'] = np.where(df_bool['signal']==1, True, False)
    df_bool['ACWI'] = np.where(df_bool['signal']==2, True, False)
    df_bool['AGG'] = np.where(df_bool['signal']==3, True, False)
    df_bool = df_bool.drop(['signal'], axis=1)
    
    # 백테스팅
    strategy_name = strategy_name
    RMTEST = bt.Strategy(
        strategy_name,
        algos = [
            bt.algos.RunMonthly(
                run_on_first_date=False,
                run_on_end_of_period=True,
                run_on_last_date=True
            ),
            # bt.algos.PrintInfo('{name}:{now}. Value:{_value:0.0f}, Price:{_price:0.4f}'),
            # bt.algos.SelectAll(),
            bt.algos.SelectWhere(df_bool),
            bt.algos.WeighEqually(),
            # bt.algos.PrintTempData(),
            bt.algos.Rebalance()
        ]
    )
    
    t = bt.Backtest(RMTEST, to_df(['SHY', 'SPY', 'ACWI', 'AGG'], start_date=start_date, end_date=end_date))
    report = bt.run(t)
    
    # 그림 그려줌?
    if plot==True:
        report.prices.plot()
    
    # HTML 만들어줌?
    if html==True:
        qs.reports.html(
            report.prices.squeeze(),
            benchmark=benchmark,
            title=strategy_name,
            download_filename='../res/' + strategy_name + '.html',
            output=True,
        )
    
    return report

start_date = '2014-01-01'
end_date = '2022-12-31'
benchmark = get_benchmark(start_date=start_date, end_date=end_date)
ODM = original_dual_momentum(strategy_name='OriginalDualMomentum', start_date=start_date, end_date=end_date, plot=True, html=True, benchmark=benchmark)
ODM.stats
```

## 7. LAA 전략

https://lazyquant.xyz/allocation/laa

```python
def LAA(strategy_name='LAA', start_date=None, end_date=None, plot=True, html=False, benchmark=None):
    US_UNEMPLOYMENT_RATE = to_df('US_UNEMPLOYMENT_RATE')
    US_UNEMPLOYMENT_RATE['US_UNEMPLOYMENT_RATE_12'] = ta.SMA(US_UNEMPLOYMENT_RATE['US_UNEMPLOYMENT_RATE'], timeperiod=12)
    SPY = to_df(['SPY'])
    SPY['SPY_200'] = ta.SMA(SPY['SPY'], timeperiod=200)
    df_m = get_end_month(pd.concat([US_UNEMPLOYMENT_RATE, SPY], axis=1).ffill()).dropna()
    signal_idx = df_m[(df_m['US_UNEMPLOYMENT_RATE']>df_m['US_UNEMPLOYMENT_RATE_12'])&(df_m['SPY']<df_m['SPY_200'])].index
    df_m['signal'] = 0
    df_m.loc[signal_idx, 'signal'] = 1

    df_bool = to_df(['IWD', 'GLD', 'IEF', 'QQQ', 'SHY'], start_date=start_date, end_date=end_date, bool_=True)
    df_bool = pd.concat([df_bool, df_m['signal']], axis=1)
    df_bool['signal'] = df_bool['signal'].ffill().bfill().astype(int)
    df_bool = df_bool.dropna()

    df_bool['IWD'] = True # 0.25
    df_bool['GLD'] = True # 0.25
    df_bool['IEF'] = True # 0.25
    df_bool['QQQ'] = np.where(df_bool['signal']==0, True, False)  # 0.25
    df_bool['SHY'] = np.where(df_bool['signal']==1, True, False)  # 0.25
    df_bool = df_bool.drop(['signal'], axis=1)

    df_weight = df_bool.div(df_bool.sum(axis=1), axis=0)
    
    # 백테스팅
    strategy_name = strategy_name
    RMTEST = bt.Strategy(
        strategy_name,
        algos = [
            bt.algos.RunMonthly(
                run_on_first_date=False,
                run_on_end_of_period=True,
                run_on_last_date=True
            ),
            # bt.algos.PrintInfo('{name}:{now}. Value:{_value:0.0f}, Price:{_price:0.4f}'),
            # bt.algos.SelectAll(),
            bt.algos.SelectWhere(df_bool),
            bt.algos.WeighTarget(df_weight),
            # bt.algos.PrintTempData(),
            bt.algos.Rebalance()
        ]
    )
    
    t = bt.Backtest(RMTEST, to_df(['IWD', 'GLD', 'IEF', 'QQQ', 'SHY'], start_date=start_date, end_date=end_date))
    report = bt.run(t)
    
    # 그림 그려줌?
    if plot==True:
        report.prices.plot()
    
    if html==True:
        qs.reports.html(
            report.prices.squeeze(),
            benchmark=benchmark,
            title=strategy_name,
            download_filename='../res/' + strategy_name + '.html',
            output=True,
        )
    
    return report

start_date = '2014-01-01'
end_date = '2022-12-31'
benchmark = get_benchmark(start_date=start_date, end_date=end_date)
LAA = LAA(strategy_name='LAA', start_date=start_date, end_date=end_date, plot=True, html=True, benchmark=benchmark)
LAA.stats
```

## 8. RAA 전략

https://almondcream.tistory.com/entry/A-%EC%9E%90%EC%82%B0%EB%B0%B0%EB%B6%84-%ED%8F%AC%ED%8A%B8%ED%8F%B4%EB%A6%AC%EC%98%A4

```python
def RAA(strategy_name='RAA', start_date=None, end_date=None, plot=True, html=False, benchmark=None):
    US_UNEMPLOYMENT_RATE = to_df('US_UNEMPLOYMENT_RATE').pct_change(12).dropna()
    US_UNEMPLOYMENT_RATE = get_bool(US_UNEMPLOYMENT_RATE)
    
    df_m = get_end_month(to_df(['VWO', 'BND'], start_date=get_date_ago_year(start_date), end_date=end_date))
    df_m = 12/1*df_m.pct_change(1) + 12/3*df_m.pct_change(3) + 12/6*df_m.pct_change(6) + 12/12*df_m.pct_change(12)
    df_m = df_m.dropna()
    df_m = get_bool(df_m)
    df_m = pd.concat([US_UNEMPLOYMENT_RATE, df_m], axis=1)
    df_m['US_UNEMPLOYMENT_RATE'] = df_m['US_UNEMPLOYMENT_RATE'].ffill()
    df_m = df_m.dropna()
    # 실업률이 1년전보다 높고(양수) VWO, BND 수익률이 하나라도 음수라면 
    df_m['signal'] = np.where((df_m['US_UNEMPLOYMENT_RATE']==True)&((df_m['VWO']==False)|(df_m['BND']==False)), True, False)
    
    df_bool = to_df(['IEF', 'TLT', 'QQQ', 'GLD', 'IEF', 'TLT', 'IWN'], bool_=True, start_date=start_date, end_date=end_date)
    df_bool.columns = ['IEF1', 'TLT1', 'QQQ', 'GLD', 'IEF2', 'TLT2', 'IWN']
    df_bool = pd.concat([df_bool, df_m['signal']], axis=1)
    df_bool['signal'] = df_bool['signal'].ffill()
    df_bool = df_bool.dropna()
    df_bool['IEF1'] = np.where(df_bool['signal']==True, True, False)
    df_bool['TLT1'] = np.where(df_bool['signal']==True, True, False)
    df_bool['QQQ'] = np.where(df_bool['signal']==False, True, False)
    df_bool['GLD'] = np.where(df_bool['signal']==False, True, False)
    df_bool['IEF2'] = np.where(df_bool['signal']==False, True, False)
    df_bool['TLT2'] = np.where(df_bool['signal']==False, True, False)
    df_bool['IWN'] = np.where(df_bool['signal']==False, True, False)
    df_bool = df_bool.drop(['signal'], axis=1)

    df_weight = df_bool.div(df_bool.sum(axis=1), axis=0)
    
    # 백테스팅
    strategy_name = strategy_name
    RMTEST = bt.Strategy(
        strategy_name,
        algos = [
            bt.algos.RunMonthly(
                run_on_first_date=False,
                run_on_end_of_period=True,
                run_on_last_date=True
            ),
            # bt.algos.PrintInfo('{name}:{now}. Value:{_value:0.0f}, Price:{_price:0.4f}'),
            # bt.algos.SelectAll(),
            bt.algos.SelectWhere(df_bool),
            bt.algos.WeighTarget(df_weight),
            # bt.algos.PrintTempData(),
            bt.algos.Rebalance()
        ]
    )
    
    df = to_df(['IEF', 'TLT', 'QQQ', 'GLD', 'IEF', 'TLT', 'IWN'], start_date=start_date, end_date=end_date)
    df.columns = ['IEF1', 'TLT1', 'QQQ', 'GLD', 'IEF2', 'TLT2', 'IWN']
    
    t = bt.Backtest(RMTEST, df)
    report = bt.run(t)
    
    # 그림 그려줌?
    if plot==True:
        report.prices.plot()
    
    if html==True:
        qs.reports.html(
            report.prices.squeeze(),
            benchmark=benchmark,
            title=strategy_name,
            download_filename='../res/' + strategy_name + '.html',
            output=True,
        )
    
    return report

start_date = '2014-01-01'
end_date = '2022-12-31'
benchmark = get_benchmark(start_date=start_date, end_date=end_date)
RAA = RAA(strategy_name='RAA', start_date=start_date, end_date=end_date, plot=True, html=True, benchmark=benchmark)
RAA.stats
```

```python
cycle_months = []
for i,j in zip(df['저-고기간'],df['고-저기간']):
    cycle_months.append(i)
    cycle_months.append(j)
cycle_months = cycle_months[2:]
```

```python
len(cycle_months)
```

```python
date='1854-12-31' 
date=datetime.strptime(date,'%Y-%m-%d')
date

cycle_list = []
cycle_list.append(date)

for i in range(len(cycle_months)):
    res = cycle_list[i]+relativedelta(months=cycle_months[i])
    cycle_list.append(res)
```
